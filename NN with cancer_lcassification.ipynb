```python
X = df.drop('benign_0__mal_1', axis=1).values #to have a numpy array
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense, Dropout
```

```python
X_train.shape
```

```python
model = Sequential ()
```

```python
model.add(Dense(30, activation = 'relu')) #30 is the number of neurons, and by repeating this line of code I just add layers
model.add(Dense(15, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid')) #because it's a binary classification problem

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test)) #i'm doing 600 epochs because I want to overfit
```

```python
losses= pd.DataFrame(model.history.history)
```

```python
losses
```

```python
losses.plot()
```

```python
#PERFECT EXAMPLE OF OVERITTING
```

```python
#based on our validation_loss, let's stop the training
#let's recreate the model
```

```python
#LET'S USE CALLBACKS: EARLY STOPPYING.
```

```python
model.add(Dense(30, activation = 'relu')) 
model.add(Dense(15, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid')) 

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
help(EarlyStopping)
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min',verbose=1, patience=25)#check the mode min, which here we want to minimaze the vol_loss
#we wait 25 epochs anyway, because there might be some noise, even if an early stop might happen earlier, we wait.
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test),
         callbacks=[early_stop]) 
```

```python
#so it stopped after 66 epochs
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

```python
#LET'S TRY NOW TO DO DROP LAYERS TECHNIQUE
```

```python
model.add(Dense(30, activation = 'relu')) 
model.add(Dropout(rate=0.5)) #rate tells how many neurons are going to be switch off (so half in this case), and they will bechosen randomly
model.add(Dense(15, activation = 'relu'))
model.add(Dropout(rate=0.5))
model.add(Dense(1, activation = 'sigmoid')) 

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
#let's use it in combination with early stopping
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test),
         callbacks=[early_stop]) 
```

```python
#it ran more now, because we were dropping out, randomly, neurons.
```

```python
model_loss=pd.DataFrame(model.history.history)
model_loss.plot()
```

```python
#that's even better!
```

```python
#we can do the predictions tasks now.
model.predict_classes(X_test)
```

```python
predictions = model.predict_classes(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```
